```python
def fixt(pk, model, fields):
    return {
        "pk": pk,
        "model": f"api.{model}",
        "fields": fields
    }
```

# USERS

```python
users = []
id_user = 1
for user in sheets.columns[:-1]:
    u_fields = {
        "name": user,
        "spotify_id": None,
        "blacklist": None,
        "friends": None
    }
    users.append(fixt(id_user, 'user', u_fields))
    id_user+=1
```
